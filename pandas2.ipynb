## Criando um DataFrame como Dicionário 

```python
dados_dicionario = {
    'Date' : ['2020-01-01', '2020-02-01', '2020-03-01', '2020-04-01', '2020-05-01'],

    'Temperatura': [29.1, 31.2, 28.5, 88.0, 24.0],

    'Classification' :['Quente', 'Muito Quente', 'Quente', 'Quente', 'Confprtavel']

}

dados_dicionario = pd.DataFrame(dados_dicionario)

display(dados_dicionario)
print(dados_dicionario)
```

## Criando DataFrame a partir de um arquivo CSV

O SCV, significa "valores separaddos por virgula". Basicamente, são arquivos de texto ondo cada linha repesenta a entrada de um valor, e essse valores são separados por um caractere delimitador, geralmente é a virgula.
Quando pasamos esse arquivo para o pandas, ele conseque intepreta esse virgulas como delitadores e organizar as informaçães em linhas e colunas corretemante

```python
display(dados_dicionario)

pd.DataFrame.to_csv(dados_dicionario, 'arquivo.csv', sep=',',index= False)
pd.DataFrame.to_csv(dados_dicionario, 'arquivo2.csv', sep=';',index= False)


```

# criar um DataFrame com valores Aleatorios do Numpy


```python
from numpy.random import randn

np.random.seed(101)

indice = [10, 20, 30, 40, 50]
colunas = ['A', 'B', 'C', 'D']
```

```python
df_aleratorio = pd.DataFrame(randn(5, 4),index= indice, columns= colunas)
display(df_aleratorio)
```

```python
display(df_aleratorio['A'])
```

```python
display(df_aleratorio[['A', 'B']])
```

```python
print(type(df_aleratorio['A']))
```

```python
df_aleratorio['nova_coluna'] = df_aleratorio['A'] + df_aleratorio['C']
```

```python
display(df_aleratorio)
```

```python
df_aleratorio.drop('nova_coluna',axis=1)
```

```python
display(df_aleratorio)
```

```python
#df_aleratorio.drop('nova_coluna',axis=1, inplace= True)
df_aleratorio = df_aleratorio.drop('nova_coluna',axis=1)
display(df_aleratorio)
```

```python
display(df_aleratorio)
```

```python
df_aleratorio.drop('A', axis= 1)
```

```python
display(df_aleratorio)

df_aleratorio.loc[10, 'B'] # Selecionando o valor da linha com indice 10 e 
```

```python
df_aleratorio.loc[20:40,['A', 'B']] # Ele esta selecionando as linha de indice 10 a ...
```
